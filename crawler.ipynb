```python
import requests
from bs4 import BeautifulSoup
from datetime import date
import dateutil.parser as dparser
```

```python
def is_minutes_html(url):
    url = str(url)
    if not url.endswith("htm"):
        return False
    if "fomcminutes" in url: #2007 - now
        return True
    elif "/fomc/minutes/" in url: #1996 - 2007
        return True
    elif "/fomc/MINUTES/" in url: #1993 - 1996
        return True
    return False # 1992 and before was pdf minutes, TODO

def construct_full_url(partial_url):
    return "https://www.federalreserve.gov" + partial_url
    

def get_recent_minutes_urls():
    urls = []
    recent_calendar_url = "https://federalreserve.gov/monetarypolicy/fomccalendars.htm"
    recent_calendar_site = requests.get(recent_calendar_url)
    parser = BeautifulSoup(recent_calendar_site.text, "html.parser")
    for potential_minutes in parser.find_all(
        "div", {"class": "col-xs-12 col-md-4 col-lg-4 fomc-meeting__minutes"}):
        for link in potential_minutes.find_all("a", href=True):
            href = link.get("href", None)
            if is_minutes_html(href):
                urls.append(construct_full_url(href))
    return urls

def get_historical_minutes_urls():
    urls = []
    historical_url_template = "https://www.federalreserve.gov/monetarypolicy/fomchistorical{}.htm"
    years = range(date.today().year, 1992, -1)
    for year in years:
        count = 0
        index_url = historical_url_template.format(str(year))
        historical_site = requests.get(index_url)
        parser = BeautifulSoup(historical_site.text, "html.parser")
        for link in parser.find_all("a", href=True):
            href = link.get("href", None)
            if is_minutes_html(href):
                urls.append(construct_full_url(href))
                count +=1
        print(f"Fetched {count} minute urls for {year}")

    return urls  
urls =  get_recent_minutes_urls() + get_historical_minutes_urls()
```

```python
for url in urls:
    date = dparser.parse(url.split("/")[-1], fuzzy=True)
    print(date, url)
```
